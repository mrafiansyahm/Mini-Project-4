```python
#Import Library
import pandas as pd
```

```python
#Import Data
df = pd.read_csv("Dataset/marketing_campaign_dataset.csv")
df.head()
```

```python
df.info()
```

```python
#Mendeteksi kolom duplikasi
df.duplicated().sum()
```

```python
#Memisahkan data numerik dan kategorikal
nums = df.select_dtypes(include=['number']).columns.tolist() #nums = numeric
cats = df.select_dtypes(exclude=['number']).columns.tolist() #cats = kategorical
```

```python
#Mendeskripsikan data
df.describe()
```

```python
#Mendeskripsikan data kategori
df.describe(include='object')
```

```python
#Melihat missing value 
df.isna().sum()
```

```python
df.shape
```

```python
#Konversi Data

#konversi kolom date
df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

```python
#Mendeteksi outlier dengan metode IQR 
Q1 = df[nums].quantile(0.25)
Q3 = df[nums].quantile(0.75)
IQR = Q3 - Q1

# Tentukan Batas Bawah dan Atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_outlier = df[((df[nums] < lower_bound) | (df[nums] > upper_bound)).any(axis=1)]
print('Data outlier: ', df_outlier.shape[0], 'baris')
```

```python
#Feature Engineering

# Menambahkan kolom CPC
df["cost_per_click"] = df["Acquisition_Cost"] / df["Clicks"]

# Menambahkan kolom CTR
df["click_trough_rate"] = df["Clicks"] / df["Impressions"]

# Menambahkan kolom CPM
df["cost_per_mile"] = (df["Acquisition_Cost"] / df["Impressions"]) * 1000

# Menambahkan kolom conversions
df["conversions"] = df["Conversion_Rate"] * df["Impressions"]

# Menambahkan kolom Ads Spend
df["ads_spend"] = df["Clicks"] * df["cost_per_click"]

# Menambahkan Income
df["Income"] = df["ads_spend"] * (1 + df["ROI"])

df.head()
```
